```python
#                             Quiz

# Создать приложение опросник, где пользователю выдается вопрос из списка
# случайным образом. На все вопросы можно ответить только правда или нет.
# То есть это True или False.

# Если пользователь отвечает правильно, то выводится некое поздравительное
# сообщение, начисляется балл и вопросник выдает следующий вопрос

# Если пользователь отвечает не правильно, то выводится другое сообщение,
# балл не начисляется и программа идет дальше. 

# По итогу выводится итоговое кол-во баллов и время, затраченное пользователем
# на прохождение опросника
```
